```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty directed graph
nervous_system = nx.DiGraph()

# Add nodes for components of the nervous system
nervous_system.add_node("Neuron")
nervous_system.add_node("Synapse")
nervous_system.add_node("Dendrite")
nervous_system.add_node("Axon")
nervous_system.add_node("Cell Body")
nervous_system.add_node("Neurotransmitter")
nervous_system.add_node("Receptor")
nervous_system.add_node("Action Potential")
nervous_system.add_node("Myelin Sheath")
nervous_system.add_node("Sensory Input")

# Add directed edges to represent causal relationships
nervous_system.add_edge("Dendrite", "Neuron")
nervous_system.add_edge("Axon", "Neuron")
nervous_system.add_edge("Cell Body", "Neuron")
nervous_system.add_edge("Synapse", "Neuron")
nervous_system.add_edge("Neurotransmitter", "Synapse")
nervous_system.add_edge("Receptor", "Synapse")
nervous_system.add_edge("Action Potential", "Axon")
nervous_system.add_edge("Myelin Sheath", "Axon")
nervous_system.add_edge("Sensory Input", "Neuron")

# Visualize the graph
nx.draw(nervous_system, with_labels=True, arrows=True)
plt.show()
```

```python
variable_names = [
    "Neuron",
    "Synapse",
    "Dendrite",
    "Axon",
    "Cell Body",
    "Neurotransmitter",
    "Receptor",
    "Action Potential",
    "Myelin Sheath",
    "Sensory Input"
]

variable_descriptions = {
    "Neuron": "A specialized cell transmitting nerve impulses.",
    "Synapse": "The junction between two neurons, allowing for signal transmission.",
    "Dendrite": "Branch-like extensions of a neuron that receive signals from other neurons.",
    "Axon": "A long fiber that transmits signals away from the neuron cell body.",
    "Cell Body": "The main part of a neuron that contains the nucleus and other cellular components.",
    "Neurotransmitter": "Chemicals released by neurons to transmit signals across synapses.",
    "Receptor": "Proteins on the surface of neurons that bind with neurotransmitters.",
    "Action Potential": "A rapid change in the electrical potential of a neuron, essential for signal transmission.",
    "Myelin Sheath": "A fatty substance that insulates and speeds up signal transmission along axons.",
    "Sensory Input": "Information received by the nervous system from sensory organs or receptors."
}
```

```python
import context_builder
```

```python
builder = context_builder.make_context()

builder.init_graph(nervous_system)
builder.observed_variables(variable_names)
builder.observed_variables_descriptions(variable_descriptions)

context = builder.build()
```

```python
ctx_params = context.get_params()
for param, value in ctx_params.items():
    if value is not None:
        print(f'The parameter {param} has this value: {value}')
```
